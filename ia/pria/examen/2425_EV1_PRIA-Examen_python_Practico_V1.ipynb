## Ejercicio 3 - Creación y manejo de una clase Alumno - 6 Puntos

Realizar un programa que conste de una clase llamada Alumno que tenga como atributos el nombre y la nota del alumno. Definir los siguientes métodos:

1. Definir la clase Alumno:

* **Atributos:**
    * nombre: nombre del alumno.
    * nota: calificación obtenida por el alumno.
* **Métodos:**
    * Método constructor: Inicializa el nombre y la nota del alumno.
    * Sobrescritura del método to_string: Devuelve una representación amigable de los atributos del alumno.
    * Método de instancia `mostrar_detalle()`: Imprime los atributos con la frase: "El alumno XXX ha sacado la nota YYY."
    * Método de clase `asignar_nota_aleatoria(cls, nombre)`: Crea un nuevo alumno con una nota aleatoria entre 0 y 10.
    * Método estático `comparar_notas(alumno1, alumno2)`: Compara las notas de dos alumnos y devuelve un mensaje indicando cuál tiene la mayor nota, o si están empatados.
2. Crear dos objetos:
    * Uno utilizando el constructor normal.
    * Otro utilizando el método de clase.
3. Utilizar los métodos:
    * Mostrar los detalles de ambos alumnos.
    * Comparar sus notas con el método estático.


```python
class Alumno:
    # constructor
    def __init__(self,nom,nota):
        self.nombre = nom
        self.nota = nota

    # sobrescribir metodo to_string
    def to_string(self):
        return 'El alumno {} ha sacado la nota {}'.format(self.nombre,self.nota)
    
    # metodo de clase 
    @classmethod
    def asignar_nota_aleatoria(cls, nombre):      
        return Alumno(nombre, np.random.randint(11))  # crea alumno con nota aleatoria entre 0 y 10
    
    # metodo estatico
    @staticmethod
    def comparar_notas(alumno1, alumno2):
        comparacion = ''
        if alumno1.nota > alumno2.nota:  comparacion = 'mayor'
        elif alumno1.nota < alumno2.nota:  comparacion = 'menor'
        else:  comparacion = 'la misma'
        return '{} tiene {} nota que {}'.format(alumno1.nombre,comparacion,alumno2.nombre)
```

```python
# crear alumnos
al1 = Alumno('Elena',7)
al2 = Alumno.asignar_nota_aleatoria('Nerea') # probar a crear con nota aleatoria

# probar to_string
print(al1.to_string())
print(al2.to_string())

# probar comparar_notas
print(Alumno.comparar_notas(al1,al2))
```

```python
# datos para grafico pie
df_wii = df.query('Name=="Wii Sports"')
wii_cols=['NA_Sales','EU_Sales', 'JP_Sales', 'Other_Sales']
wii_conteo = []
for i,row in df_wii.iterrows():
    wii_conteo.append(row['NA_Sales'])
    wii_conteo.append(row['EU_Sales'])
    wii_conteo.append(row['JP_Sales'])
    wii_conteo.append(row['Other_Sales'])


# datos para grafico bar
df_top5 = df.sort_values(by='Global_Sales', ascending=True)
df_top5 = df_top5[:5] # primeros 5
cols_top5 = df_top5['Platform']
conteo_top5 = []  
for i,row in df_top5.iterrows():
    wii_conteo.append(row['Global_Sales'])
```

```python
# Creamos el objeto figure
fig = plt.figure(constrained_layout = True)

# Configuramos que tendrá 3 filas y 3 columnas
gs = fig.add_gridspec(3, 3)


## Crear subplots 
# Ocupa todas las todas las filas salvo la última y todas las columnas
fig_ax1 = fig.add_subplot(gs[:-1, :])
fig_ax1.pie(wii_conteo, labels = wii_cols ,
         colors = ["#fbb4ae", "#b3cde3", "#ccebc5", "#decbe4"], shadow = True,
         startangle = 90, explode = [0,0,0.5,0],  autopct='%1.1f%%')
fig_ax1.set_title("Wii Sports")


# Ocupa todas la última filas y primeras 2columnas
fig_ax2 = fig.add_subplot(gs[-1:, :2])
# sns.scatterplot(data = df, x = "NA_Sales", y = "EU_Sales", palette= 'Set3',size = "JP_Sales")
fig_ax2.scatter(x = df['NA_Sales'], y = df['EU_Sales'], cmap ='Set3' ,s=50)
fig_ax2.set_title("Ventas")
# plt.colorbar()


# Ocupa la última fila y la última columna
fig_ax3 = fig.add_subplot(gs[-1, -1])
fig_ax3.bar(cols_top5, conteo_top5, color = "#809ab0")
fig_ax3.set_title("top 5 ventas globales")

plt.show()
```
